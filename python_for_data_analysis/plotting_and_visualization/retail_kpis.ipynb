```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Volume of transactions**

```python
plt.figure(figsize=(9, 6))

x = np.arange(4)
volume_k = [12, 34, 31, 20]

plt.bar(x, volume_k, width=0.4, color='#177cb0')
plt.tick_params(labelsize=13)
plt.xticks(np.arange(4), ('apple', 'banana', 'cherry', 'durian'))
plt.xlabel('Product', size=13)
plt.ylabel('Volume (k)', size=13)

plt.show()
```

This plot describes the volume(k) for each fruit. Among four fruits, bananas’ sales are the largest (34k units), however, it seems that consumers don’t like apple that much.

**Turnover of transactions**

```python
plt.figure(figsize=(9, 6))

year_n_1 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]
year_n = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]

plt.plot(np.arange(12), year_n_1, marker='.', color='#177cb0',
         label='year N-1', linestyle='--', alpha=0.5)
plt.plot(np.arange(12), year_n, marker='.', color='#177cb0', label='year N')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(12), np.arange(1,13))
plt.xlabel('Month', size=13)
plt.ylabel('Turnover (K euros) of ice-cream', size=13)
plt.ylim(bottom=0)
plt.legend(fontsize=13)

plt.show()
```

This plot compares two-year turnover together. The peak for both year is in the summer, for year N, the peak is in August; however, the peak is reached in June in the year N-1, which is caused by the heatwave in June.

**Transactions amount**

```python
plt.figure(figsize=(9, 6))

trx_nb = [0, 5, 15, 13, 10, 14, 20, 22, 13, 15, 12, 18, 23, 20, 16, 12, 3, 0]
plt.plot(np.arange(7, 25), trx_nb, marker='.', color='#177cb0')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(7, 25), np.arange(7, 25))
plt.xlabel('Hour', size=13)
plt.ylabel('Transaction amount (K)', size=13)
plt.ylim(bottom=0)

plt.show()

```

According to this plot, it's obvious that the peak in a day is around noon and afterworking like 19h, because customers work in the morning and afternoon, they have time to shopping during the noon and afterworking. Moreover, there are 2 small peak around 9h and 16h, that might because customer come to buy their breakfast and classes over around 16h.

**Average volume per transaction**

```python
plt.figure(figsize=(9, 6))

vol_per_trx = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]

plt.bar(np.arange(12), vol_per_trx, width=0.4, color='#177cb0')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(12), np.arange(1,13))
plt.xlabel('Month', size=13)
plt.ylabel('Average volume of ice-cream\nper transaction', size=13)
plt.ylim(bottom=0)

plt.show()
```

This plot show the average volume per transaction of ice-cream. Logically, the volume increases with temperature's increasing.

**Average purchase value**

```python
fig, axarr = plt.subplots(nrows=1, ncols=1, figsize=(9, 6))

vol_per_trx = [8.5, 7, 10, 11, 10, 12.3, 13.2, 11.6, 13, 14, 12, 11]
val_per_trx = [20, 25, 30, 25, 33, 27, 24, 26, 25, 28, 20, 22]

vol_bar = axarr.bar(np.arange(12), vol_per_trx, width=0.4, color='#177cb0')

axarr_2 = axarr.twinx()
val_line = axarr_2.plot(np.arange(12), val_per_trx, marker='.', color='#FF9415')

axarr.tick_params(labelsize=12)
axarr.set_xticks(np.arange(12))
axarr.set_xticklabels(np.arange(1, 13))
axarr.set_xlabel('Month', size=13)
axarr.set_ylabel('Average volume per transaction\n(k units)', size=13)
axarr.set_ylim(bottom=0, top=20)

axarr_2.tick_params(labelsize=12)
axarr_2.set_ylim(bottom=0, top=35)
axarr_2.set_ylabel('Average value per transaction\n(k euros)', size=13, rotation=270, labelpad=33)
axarr_2.legend((vol_bar[0], val_line[0]), ('Volume', 'Turnover'))

plt.show()
```

This plot describes the average volume and average value per sale during 12 months. We find that when the average volume per sale increases, the average value does not always increase. For instance, the average volume of April is more than March, but the average value is less, that means customers buy more units during their purchase, but the price per unit is decreasing.
